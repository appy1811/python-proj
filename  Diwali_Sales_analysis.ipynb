# Exploratory Data Analysis

## Gender

```python
sns.countplot(x='Gender',data = df,hue='Gender' ,palette = "Blues_d")
```

```python
test = sns.countplot(x='Gender', data = df,)
for bars in test.containers:
    test.bar_label(bars)
```

```python
sales_gen = df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Gender',y='Amount',data = sales_gen, hue='Gender' ,palette = "Blues_d")
```

###### From above graph we can see that female buyers are more than male also female spends most amount.

## Age

```python
sns.countplot(x='Age Group',data = df, hue='Gender' )
```

```python
test = sns.countplot(x='Age Group',data = df, hue='Gender')
for bars in test.containers:
    test.bar_label(bars)
```

```python
sales_age = df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Age Group',y='Amount',data = sales_age, hue='Age Group')
```

##### From above graph we can see that most of the buyers are from age group between 26-35 and mostly female.

#### State

```python
#Total number of orders from top 10 States
sales_state = df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data=sales_state, x='State', y ='Orders',hue='State')
```

```python
#Total amount sales from top 10 states
sales_state = df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(17,5)})
sns.barplot(data=sales_state, x='State', y ='Amount',hue='State')
```

###### from above graph we can see that ost of the orders & total sales/amount are from UP,MH and Karnataka 

### Marital status

```python
test = sns.countplot(x='Marital_Status',data = df, hue='Marital_Status' )
for bars in test.containers:
    test.bar_label(bars)
```

```python
sales_status = df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(data = sales_status, x = 'Marital_Status', y='Amount', hue='Gender')
```

##### From above graph we can see that most of the buyers are married women and they have high purchasing power

### Occupation

```python
sns.set(rc={'figure.figsize':(20,5)})
test = sns.countplot(data = df, x = 'Occupation' )

for bars in test.containers:
    test.bar_label(bars)
```

```python
sales_occ = df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_occ, x = 'Occupation', y='Amount', hue='Occupation')
```

#### we can see that most of the buyers are from the IT sector

### Product Category

```python
sns.set(rc={'figure.figsize':(20,5)})
test = sns.countplot(data = df, x = 'Product_Category' ,hue='Product_Category')

for bars in test.containers:
    test.bar_label(bars)
```

```python
sales_prod = df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_prod, x = 'Product_Category', y='Amount', hue='Product_Category')
```

##### from above graph we can see that most of the sold products are from food,footwear and Electronics

```python
sales_prod = df.groupby(['Product_ID'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_prod, x = 'Product_ID', y='Orders', hue='Product_ID')
```

##### Married women from age group 26-35 yrs from UP, MH and Karnataka working in IT sector,Healthcare & aviation are more likely to buy products from food, clothing and electronics.
